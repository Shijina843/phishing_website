```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings, json, pickle
from pathlib import Path

warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid', palette='muted', font_scale=1.1)

BASE   = Path('../..')
DATA   = BASE / 'DataFiles'
MODELS = BASE / 'models'
RESULT_IMAGES = BASE / 'Result image'

print('Ready ✓')
```

```python

df = pd.read_csv(DATA / 'features.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Class distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

counts = df['label'].value_counts()
colors = ['#2ecc71', '#e74c3c']
axes[0].pie(counts, labels=['Legitimate','Phishing'], colors=colors,
            autopct='%1.1f%%', startangle=90, textprops={'fontsize':12})
axes[0].set_title('Class Distribution', fontsize=14, fontweight='bold')

sns.countplot(x='label', data=df, palette={'0':'#2ecc71','1':'#e74c3c'},
              ax=axes[1])
axes[1].set_xticklabels(['Legitimate','Phishing'])
axes[1].set_title('URL Count by Class', fontsize=14, fontweight='bold')
axes[1].set_xlabel('')
for p in axes[1].patches:
    axes[1].annotate(f'{int(p.get_height()):,}',
                     (p.get_x()+p.get_width()/2., p.get_height()),
                     ha='center', va='bottom', fontsize=12)
plt.tight_layout()
plt.savefig(RESULT_IMAGES / 'class_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 2. Feature Analysis

```python
feature_cols = [c for c in df.columns if c != 'label']

# Feature means by class
means = df.groupby('label')[feature_cols].mean()
diff  = (means.loc[1] - means.loc[0]).sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(12, 6))
colors = ['#e74c3c' if v > 0 else '#2ecc71' for v in diff.values]
ax.barh(diff.index, diff.values, color=colors)
ax.axvline(0, color='black', linewidth=0.8)
ax.set_xlabel('Mean Difference (Phishing − Legitimate)', fontsize=12)
ax.set_title('Feature Discriminative Power\n(Red = skewed toward phishing, Green = legitimate)', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(RESULT_IMAGES / 'feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Correlation heatmap
fig, ax = plt.subplots(figsize=(14, 10))
mask = np.triu(np.ones_like(df[feature_cols].corr(), dtype=bool))
sns.heatmap(df[feature_cols].corr(), mask=mask, annot=True, fmt='.2f',
            cmap='RdBu_r', center=0, linewidths=0.5, ax=ax,
            annot_kws={'size': 8})
ax.set_title('Feature Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(RESULT_IMAGES / 'correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. Model Performance Comparison

```python
with open(MODELS / 'metrics.json') as f:
    metrics = json.load(f)

metrics_df = pd.DataFrame(metrics)
print(metrics_df[['model','accuracy','precision','recall','f1_score','roc_auc']].to_string(index=False))
```

```python
# Grouped bar chart
cols  = ['accuracy','precision','recall','f1_score','roc_auc']
x     = np.arange(len(metrics_df))
width = 0.15
palette = ['#3498db','#e74c3c','#2ecc71','#f39c12','#9b59b6']

fig, ax = plt.subplots(figsize=(14, 6))
for i, (col, clr) in enumerate(zip(cols, palette)):
    ax.bar(x + i*width, metrics_df[col], width, label=col.replace('_',' ').title(), color=clr)

ax.set_xticks(x + width * 2)
ax.set_xticklabels(metrics_df['model'], rotation=20, ha='right')
ax.set_ylim(0, 105)
ax.set_ylabel('Score (%)')
ax.set_title('Model Performance Comparison', fontsize=14, fontweight='bold')
ax.legend(loc='lower right')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda v,_: f'{v:.0f}%'))
plt.tight_layout()
plt.savefig(RESULT_IMAGES / 'model_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Confusion matrices
fig, axes = plt.subplots(1, len(metrics), figsize=(18, 4))
for ax, m in zip(axes, metrics):
    cm = np.array(m['confusion_matrix'])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax,
                xticklabels=['Legit','Phish'],
                yticklabels=['Legit','Phish'])
    ax.set_title(m['model'], fontsize=10, fontweight='bold')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')
plt.suptitle('Confusion Matrices — All Models', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig(RESULT_IMAGES / 'confusion_matrices.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Feature Importance (Random Forest)

```python
with open(MODELS / 'random_forest.pkl', 'rb') as f:
    rf = pickle.load(f)

importances = pd.Series(rf.feature_importances_, index=feature_cols).sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(10, 7))
colors = plt.cm.RdYlGn(np.linspace(0.2, 0.9, len(importances)))
importances.plot.barh(color=colors, ax=ax)
ax.set_xlabel('Feature Importance Score')
ax.set_title('Random Forest — Feature Importances', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(RESULT_IMAGES / 'rf_feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()
```
